```python
import xlsxwriter
from Bio.Blast import NCBIXML

import pandas as pd
i=16000
name1="res_"+str(i)+".xml"
name2="seqs_"+str(i)+".txt"
name3=str(i)+"_df"+".xlsx"
#open the results file
result=open(name1,"r")
records= NCBIXML.parse(result)
item=records

trial_3=pd.DataFrame()

alignments=[]
start=[]
end=[]
linker=[]
matched_indexes=[]
qid=[]           
allq=[]
defn=[]

seqlist=[]
s=[]
f=open(name2, "r")
ksj= 1
for line in f.readlines():
    if ksj % 2 == 0 :
        seqlist.append(line.rstrip())
    ksj += 1

#list1.pop(0)
ks=0
trialdf=pd.DataFrame()
trial_2=pd.DataFrame()
for record in records:

    allq.append(record.query)
    for alignment in record.alignments:
        for hsp in alignment.hsps:
            start=[]
            #only first hsp is taken for each hit, this is the hsp with the best score
            if alignment.hit_def:
                start.append(seqlist[ks])
            
            start.append(hsp.query_start)
            start.append(hsp.query_end)
            start.append(alignment.hit_def)
            allq.append("   ")
            #trialdf=pd.DataFrame({"Sequence":pd.Series(start[0]), "Start":pd.Series(start[1]),"End":pd.Series(start[2]),"Alignment Definition":pd.Series(start[3])})
            trialdf=pd.DataFrame({"Sequence":pd.Series(start[0]), "Start":pd.Series(start[1]),"End":pd.Series(start[2]),"Alignment Definition":pd.Series(start[3])})
            trial_2=pd.concat([trial_2,trialdf], axis=0)
            
            break
    ks=ks+1
    #print(allq)
    allq.pop()


trial_2.insert(0,"protein",allq)
trial_2.to_excel(name3, index=False)    
#check for any anomaly
#print(len(start))
#print(len(end))
#print(len(alignments))

        

```
